Q2.Use the Linear Regression Example Code to keep math score as X and "reading score as Y

```
import numpy as np
from sklearn.linear_model import LinearRegression


X= np.array(stats.iloc[:,5].values)
y = np.array(stats.iloc[:,6].values)

X= X.reshape(-1, 1)
y= y.reshape(-1, 1)

print( 'X' , X [:5] )
print( 'y' , y [:5] )
```

```
model = LinearRegression()
model.fit( X , y )
model.score( X , y )
```
